```python
# Clase Mucama
class Mucama:
    # Constructor (atributos)
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Funciones (acciones)
    def barrer(self):
        return f"{self.nombre} está barriendo el piso."

    def trapear(self):
        return f"{self.nombre} está trapeando el piso."

    def tender_camas(self):
        return f"{self.nombre} está tendiendo las camas."


# Creación de objetos (instancias)
sirvienta1 = Mucama("María López", "6561234567", 35)
sirvienta2 = Mucama("Ana García", "6569876543", 28)

# Uso de los objetos
print("=== Datos de Sirvienta 1 ===")
print("Nombre:", sirvienta1.nombre)
print("Teléfono:", sirvienta1.telefono)
print("Edad:", sirvienta1.edad)
print(sirvienta1.barrer())
print(sirvienta1.trapear())
print(sirvienta1.tender_camas())

print("\n=== Datos de Sirvienta 2 ===")
print("Nombre:", sirvienta2.nombre)
print("Teléfono:", sirvienta2.telefono)
print("Edad:", sirvienta2.edad)
print(sirvienta2.barrer())
print(sirvienta2.trapear())
print(sirvienta2.tender_camas())

print("El programa ah terminado")
print("Programa generado por Arevalo Rafael NL = 3")
```
